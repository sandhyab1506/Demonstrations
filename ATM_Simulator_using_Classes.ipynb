```python

#*******************************************************************************************************************************************************************************************************************************************
#
#   ATM Simulator using Clasees Project by Sandhya Buddharaju
#   Language : Python
#   Version 1.0
#
#
#
#*****************************************************************************************************************************************************************************************************************************************************

class ATM:

  #attribute declarations

  # main menu items
  main_menu = ["Display Balance","Deposit Money","Withdraw Money","Exit"]

  # default set ATM pin to 1234
  atm_pin = 1234

  # default set Current Balance to £100
  current_balance = 100


  # Fuction to set current balance
  @classmethod
  def set_current_balance(self,balance):
    self.current_balance = balance

  # Fuction to set atm pin balance
  @classmethod
  def set_atm_pin(self,pin):
    self.atm_pin = pin

  # Fuction to get current balance
  @classmethod
  def get_current_balance(self):
    return self.current_balance

  # Fuction to get atm pin
  @classmethod
  def get_atm_pin(self):
    return self.atm_pin

  # Function to display Main Menu for user selection
  @classmethod
  def print_main_menu(self):
    print("\n********ATM Menu*******\n")
    for item in self.main_menu:
      print(str(self.main_menu.index(item)+1)+") "+item)

  # Function to print the balance
  @classmethod
  def print_balance(self):
    print("Current Balance : £",self.current_balance,".")

  @classmethod
  def deposit_money(self,deposit_amount):
    self.current_balance = self.current_balance + deposit_amount
    print("Deposit of £",deposit_amount," accepted.")

  @classmethod
  def withdraw_money(self,withdraw_amount):
    self.current_balance = self.current_balance - withdraw_amount
    print("Please collect your cash £",withdraw_amount,".")




# create ATM class object
xbank_atm_obj = ATM()

# pre set ATM pin to 0000 using object method
xbank_atm_obj.set_atm_pin(1111)

# pre set Current Balance to £500 using object method
xbank_atm_obj.set_current_balance(500)


def continue_option():
    option = input("\nDo you wish to continue Y/N :")
    if(option == 'Y' or option == 'y'):
      to_continue = True
      xbank_atm_obj.print_main_menu()
    elif(option == 'N' or option == 'n'):
      print("\n\nThank You for using X Bank today. Have a Great Day Ahead.")
      to_continue = False
    else:
      print(" Sorry input not recognized. Thank you for using X Bank Today. Have a Great Day Ahead.")
      to_continue = False
    return to_continue

# get pin from user and verify against stored value
user_atm_pin = int(input("Please Enter Pin :"))
to_continue = True
is_exit = True

while(is_exit):
  if(user_atm_pin == xbank_atm_obj.get_atm_pin()):
      xbank_atm_obj.print_main_menu()
      while(to_continue):
        option_selection = int(input("\nPlease enter option 1/2/3/4 : "))
        # option 1 -- Print Balance on Screen
        if(option_selection == 1):
          xbank_atm_obj.print_balance()
          to_continue = continue_option()
        # option 2 -- Get the deposit amount and add to current balnace and Print Balance on Screen
        elif(option_selection == 2):
          deposit_amount = int(input("Please enter amount to deposit : £"))
          xbank_atm_obj.deposit_money(deposit_amount)
          xbank_atm_obj.print_balance()
          to_continue = continue_option()
        # option 3 -- Get the withdrwal amount and subtract from current balnace and Print Balance on Screen
        elif(option_selection == 3):
          xbank_atm_obj.print_balance()
          withdraw_amount = int(input("Please enter amount to withdraw : £"))
          if(xbank_atm_obj.get_current_balance() >= withdraw_amount):
            xbank_atm_obj.withdraw_money(withdraw_amount)
            xbank_atm_obj.print_balance()
          else:
            print("Sorry you do not have sufficient funds in your Account to withdraw £",withdraw_amount,"\n Please enter any amount less than £", xbank_atm_obj.get_current_balance())
          to_continue = continue_option()
        # option 4 -- Exit
        elif(option_selection == 4):
          print("Thank You for using x Bank today. Have a Great Day Ahead.")
          to_continue = False
          is_exit = False
        else:
          print("Please select valid option! ")
        is_exit=False
  else:
    user_atm_pin = int(input("PIN number incorrect. Please enter valid PIN:"))

```
